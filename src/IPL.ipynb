```python
import pandas as pd
import numpy as np
import os
import tqdm
```

```python
data = pd.read_csv("../Data/ipl_csv2/all_matches.csv")
```

```python
data
```

```python
data.columns
```

```python
fill_col_0 = ['wides', 'noballs', 'byes', 'legbyes', 'penalty']
fill_col_none = ['wicket_type', 'player_dismissed']
```

```python
data[fill_col_0] = data[fill_col_0].fillna(0)
data[fill_col_none] = data[fill_col_none].fillna("none")
```

```python
data = data.drop(['other_wicket_type', 'other_player_dismissed'], axis=1)
```

```python
data
```

```python
from tqdm import tqdm
```

```python
%timeit

data['total'] = np.zeros(data.shape[0])
data['wickets'] = np.zeros(data.shape[0])
batsman = {}
team_bat ={}
team_bowl ={}
venue = {}

for i in tqdm(range(data.shape[0])):
    
    striker = data['striker'].iloc[i].lower()
    flag = batsman.get(striker, None)
    runs_in_a_ball = data['runs_off_bat'].iloc[i]
    wicket_of_batsman = 1 if data['wicket_type'].iloc[i] != 'none' else 0
    matchid = data['match_id'].iloc[i]
    
    if not flag:
        batsman[striker] = {'ball_count': 1, 'run_count': runs_in_a_ball, 'innings_count':1, 'innings_out_count': 0 ,\
                             'matchid': matchid}
    else:
        i_c = 1 if flag['matchid'] != matchid else 0
        batsman[striker] = {'ball_count': flag['ball_count']+1, 'run_count': flag['run_count']+runs_in_a_ball,\
                              'innings_count':flag['innings_count'] + i_c, 'innings_out_count':flag['innings_out_count'], 'matchid': matchid}
    
    
    player_dism = data['player_dismissed'].iloc[i].lower()
    player_out = batsman.get(player_dism, None)
    
    if not player_out:
        batsman[player_dism] = {'ball_count': 0, 'run_count': 0, 'innings_count':1, 'innings_out_count':1,\
                             'matchid': matchid}
    else:
        batsman[player_dism]['innings_out_count'] = batsman[player_dism]['innings_out_count'] + 1 
        
        i_cc = 1 if player_out['matchid'] != matchid else 0
        batsman[player_dism]['innings_count'] = batsman[player_dism]['innings_count'] + i_cc
        batsman[player_dism]['matchid'] = matchid
    
    
    
    if data['ball'].iloc[i] == 0.1: 
        data['total'].iloc[i] = data['runs_off_bat'].iloc[i] + data['extras'].iloc[i]  
        
        
        if data['wicket_type'].iloc[i] != 'none':
            data['wickets'].iloc[i] = 1
        else:
            data['wickets'].iloc[i] = 0
            
        
    else :
        data['total'].iloc[i] = data['total'].iloc[i-1] + data['runs_off_bat'].iloc[i] + data['extras'].iloc[i]
        
        
        if data['wicket_type'].iloc[i] != 'none':
            data['wickets'].iloc[i] = data['wickets'].iloc[i-1] + 1
        else:
            data['wickets'].iloc[i] = data['wickets'].iloc[i-1]
    
    
    bat_team = data['batting_team'].iloc[i]
    flag2 = team_bat.get(bat_team, None)
    
    run_scored = data['runs_off_bat'].iloc[i] + data['extras'].iloc[i]
    
    if not flag2:
        team_bat[bat_team] = {'runs':run_scored,'match_count' : 1,'match_id' : matchid, 'wickets' : wicket_of_batsman }
    
    else:
        team_bat[bat_team]['runs'] = flag2['runs'] + run_scored
        i_cc2 = 1 if flag2['match_id'] != matchid else 0
        team_bat[bat_team]['match_count'] = i_cc2 + flag2['match_count']
        team_bat[bat_team]['match_id'] = matchid
        team_bat[bat_team]['wickets'] = flag2['wickets'] + wicket_of_batsman
    
    if data['ball'].iloc[i]== 7.1:
        team_bat[bat_team]['6_over_score'] = team_bat[bat_team].get('6_over_score',0) + data['total'].iloc[i] - run_scored
        team_bat[bat_team]['6_over_wickets'] = team_bat[bat_team].get('6_over_wickets',0) + data['wickets'].iloc[i] - wicket_of_batsman
    
    bowl_team = data['bowling_team'].iloc[i]
    flag3 = team_bowl.get(bowl_team, None)
    
    if not flag3:
        team_bowl[bowl_team] = {'runs':run_scored,'match_count' : 1,'match_id' : matchid, 'wickets' : wicket_of_batsman }

    else:
        team_bowl[bowl_team]['runs'] = flag3['runs'] + run_scored
        i_cc3 = 1 if flag3['match_id'] != matchid else 0
        team_bowl[bowl_team]['match_count'] = i_cc3 + flag3['match_count']
        team_bowl[bowl_team]['match_id'] = matchid
        team_bowl[bowl_team]['wickets'] = flag3['wickets'] + wicket_of_batsman
    
    if data['ball'].iloc[i]== 7.1:
        team_bowl[bowl_team]['6_over_score'] = team_bowl[bowl_team].get('6_over_score',0) + data['total'].iloc[i] - run_scored
        team_bowl[bowl_team]['6_over_wickets'] = team_bowl[bowl_team].get('6_over_wickets',0) + data['wickets'].iloc[i] - wicket_of_batsman
    
    if data['ball'].iloc[i] == 7.1:
        
        if not venue.get(data['venue'].iloc[i], None):
            venue[data['venue'].iloc[i]] = {'6_over_run' : data['total'].iloc[i] - run_scored, \
                                       '6_over_wickets' : data['wickets'].iloc[i] - wicket_of_batsman, 'match_id' : matchid,\
                                       'match_count' : 1}
        
        else:
            i_cc4 = 1 if venue[data['venue'].iloc[i]].get('match_id') != matchid else 0
            venue[data['venue'].iloc[i]] = {'6_over_run' : venue[data['venue'].iloc[i]].get('6_over_run',0) + data['total'].iloc[i] - run_scored, \
                                       '6_over_wickets' : venue[data['venue'].iloc[i]].get('6_over_wickets',0) + \
                                        data['wickets'].iloc[i] - wicket_of_batsman, 'match_id' : matchid,\
                                       'match_count' : venue[data['venue'].iloc[i]].get('match_count',0) + i_cc4 }
    
    
```

```python
for k in batsman.keys():
    if batsman[k]['run_count'] != 0:
        batsman[k]['strike_rate'] = batsman[k]['run_count']/batsman[k]['ball_count']
    else:
        batsman[k]['strike_rate'] = 0
    
    if batsman[k]['innings_out_count'] != 0:
        batsman[k]['average'] = batsman[k]['run_count']/batsman[k]['innings_out_count']
    else:
        batsman[k]['average'] = batsman[k]['run_count']
```

```python
for k in team_bat.keys():
    team_bat[k]['6_over_avg_score'] = team_bat[k]['6_over_score']/team_bat[k]['match_count']
    team_bat[k]['6_over_avg_wickets'] = team_bat[k]['6_over_wickets']/team_bat[k]['match_count']
     
```

```python
for k in team_bowl.keys():
    team_bowl[k]['6_over_avg_score'] = team_bowl[k]['6_over_score']/team_bowl[k]['match_count']
    team_bowl[k]['6_over_avg_wickets'] = team_bowl[k]['6_over_wickets']/team_bowl[k]['match_count']
     
```

```python
for k in venue.keys():
    venue[k]['6_over_avg_score'] = venue[k]['6_over_run']/(2*venue[k]['match_count'])
    venue[k]['6_over_avg_wickets'] = venue[k]['6_over_wickets']/(2*venue[k]['match_count'])
     
```
